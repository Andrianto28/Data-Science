```python
import pandas as pd
import numpy as np
import random

data = {'feature_1': [random.uniform(0,1) for i in range (150)],
        'feature_2': [random.uniform(0,1) for i in range (150)],
        'feature_3': [random.uniform(0,1) for i in range (150)],
        'feature_4': [random.uniform(0,1) for i in range (150)],
        'feature_5': [random.uniform(0,1) for i in range (150)],
        'feature_6': [random.uniform(0,1) for i in range (150)],
        'feature_7': [random.uniform(0,1) for i in range (150)],
        'feature_8': [random.uniform(0,1) for i in range (150)],
        'feature_9': [random.uniform(0,1) for i in range (150)],
        'feature_10': [random.uniform(0,1) for i in range (150)],
        'feature_11': [random.uniform(0,1) for i in range (150)],
        'feature_12': [random.uniform(0,1) for i in range (150)],
        'feature_13': [random.uniform(0,1) for i in range (150)],
        'feature_14': [random.uniform(0,1) for i in range (150)],
        'feature_15': [random.uniform(0,1) for i in range (150)],
        'label': [random.choice(['A', 'B', 'C', 'D']) for i in range (150)]}
df = pd.DataFrame(data)
df.head()
```

```python
from sklearn.decomposition import PCA
#mendefinisika jumlah kolom baru yg diinginkan
# type code here
pca = PCA(n_components=2)
```

```python
# Type Code Here
fit_pca = pca.fit_transform(x)
```

```python

# Type Code Here
pca_df = pd.DataFrame(data = fit_pca, columns = ['PCA_1', 'PCA_2'])
pca_df.tail()
```

```python
finalDf = pd.concat([pca_df, df[['label']]], axis = 1)
finalDf.head(5)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)


targets = ['A', 'B', 'C','D']
colors = ['r', 'g', 'b','y']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['label'] == target
    ax.scatter (finalDf.loc[indicesToKeep, 'PCA_1']
               , finalDf.loc[indicesToKeep, 'PCA_2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
#code latihan 2 (LDA)
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis 
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#load iris dataset
iris = datasets.load_iris()
#convert dataset to pandas DataFrame
df = pd.DataFrame(data = np.c_[iris['data'], iris['target']],
                 columns = iris['feature_names'] + ['target'])
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names) #menambahkan kolom kategorik
df.columns = ['s_length', 's_width', 'p_length', 'p_width', 'target', 'species'] #Menggantik nama nama kolom

#view first six rows of DataFrame
df.head()
```

```python
iris['feature_names']
```

```python
iris.feature_names
```

```python
#mencari seberapa banyak data yg ada di dataframe
len(df.index)
```

```python
#define predictor and response variables
X = df[['s_length', 's_width', 'p_length', 'p_width']] #perintah ini sama dengan df.iloc[:, :4]
y = df['species'] #perintah sama saja dengan df.iloc[:,-1]

#Fit the LDA model
model = LinearDiscriminantAnalysis()
model.fit(X, y)
```

```python
df.iloc[:,-1]
```

```python
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1) # n_split jumlah pembagian data, 
#n_repeat seberapa banyak data akan diulang, random state yg akan mengontrol perubahan pada data

#evaluate model
scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
print(np.mean(scores)) 
```

```python
#define new observation
new = [5, 3, 1, .4]

#predict which class the new observation belongs to
model.predict([new])
```
